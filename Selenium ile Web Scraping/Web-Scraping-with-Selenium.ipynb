```python
import selenium
from selenium import webdriver
from selenium.webdriver.chrome.service import Service


path = "C:\Program Files (x86)\chromedriver.exe"

# ChromeDriver servisini oluturun
service = Service(executable_path=path)

# WebDriver' balatn
driver = webdriver.Chrome(service=service)

# ekilecek verinin adresi
driver.get("https://www.cimri.com/cep-telefonlari/en-ucuz-samsung-galaxy-s20-fe-128gb-6gb-ram-6-5-inc-12mp-akilli-cep-telefonu-fiyatlari,a633775428")

```

```python
#veri ekileek blmn html kodlarna eriip copy> copy xpath kodunu kopyalyoruz
# driver.find_element("xpath","//*[@id='__next']/main/div[1]/div[5]")
elements = driver.find_element("xpath","//*[@id='__next']/main/div[1]/div[5]")
new = elements.text
new
```

```python
#verileri \n e gre ayrdk  ve dataframe ile tablolatrdk

df=pd.DataFrame(new.split("\n"))

#tablo eklindeki verileri dzenleme


#batan indexliyor elenenler ile
dff =df.iloc[10:].reset_index(drop= True)
dff = dff.rename(columns = {0: "columns"})
dff
```

```python
# Dzenlenen veri erevesini bir CSV dosyasna aktarma
dff.to_csv('output.csv', index=False, header=True)
```
